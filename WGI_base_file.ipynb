```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
import csv
import os
import numpy as np
from chardet import detect
```

```python
# Store filepath in a variable
VAA_file = "Resources/VoiceandAccountability.csv"
ROL_file = "Resources/RuleofLaw.csv"
RQ_file = "Resources/RegulatoryQuality.csv"
PSNV_file = "Resources/PoliticalStabilityNoViolence.csv"
GE_file = "Resources/GovernmentEffectiveness.csv"
COC_file = "Resources/ControlofCorruption.csv"
```

```python
final_VAC_df = renamed_VAC_df.drop([0])
final_VAC_df.head()
```

```python
final_ROL_df = renamed_ROL_df.drop([0])
final_ROL_df.head()
```

```python
COC_df = pd.read_csv(COC_file, encoding='ISO-8859-1', delimiter=',')
COC_df.head()
```

```python
renamed_COC_df = COC_df.rename(columns={"Unnamed: 0":"Country/Territory", "Unnamed: 1":"WBCode"})
```

```python
final_COC_df = renamed_COC_df.drop([0])
final_COC_df.head()
```

```python
VAC_ROL_df = pd.merge(final_VAC_df, final_ROL_df, how='outer', on='Country/Territory')
VAC_ROL_df.head()
```

```python
RQ_PSNV_df = pd.merge(final_RQ_df, final_PSNV_df, how='outer', on='Country/Territory')
RQ_PSNV_df.head()
```

```python
GE_COC_df = pd.merge(final_GE_df, final_COC_df, how='outer', on='Country/Territory')
GE_COC_df.head()
```

```python
four_df = pd.merge(VAC_ROL_df, RQ_PSNV_df, how='outer', on='Country/Territory')
four_df.head()
```

```python
six_df = pd.merge(four_df, GE_COC_df, how='outer', on='Country/Territory')
six_df.head()
```

```python
all_WGI_df = six_df.rename(columns={"WBCode_x_x":"Country Code"})
all_WGI_df.head()
```

```python
list(all_WGI_df.columns.values)
```
